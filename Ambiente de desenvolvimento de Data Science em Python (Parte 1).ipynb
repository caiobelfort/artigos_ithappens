## 2. IPython 

```python
Image('http://ipython.readthedocs.io/en/stable/_images/ipython-6-screenshot.png')
```

Quando instalamos o **Python**, normalmente estamos obtendo o interpretador da linguagem (utilizado para executar os scripts) e as bibliotecas padrões da linguagem. Apesar desse ambiente já ser bem completo para muitos projetos, algumas pessoas achavam que faltava uma certa interatividade para o interpretador e foi ai que surgiu o **IPython** (Interactive Python).

Basicamente é um projeto que coloca esteróides no interpredador padrão do Python, tornando possível usar o *autocomplete* com um simples TAB, plotar gráficos *inline* (sem necessidade de salvar e abrir a imagem fora do ambiente do interpretador), executar comandos *bash* como *ls, grep, etc.* e meus prefiridos que costumamos chamar de *IPython magics* que são pequenas macros que auxiliam no desenvolvimento do trabalho.

Abaixo temos um exemplo comparando a implementação comum para medir o tempo de execução de um código com o interpretador padrão e como é feita a medição dentro do IPython.

```python
# Exemplo básico de medição de tempo de execução encontrado em https://docs.python.org/3/library/timeit.html
import timeit

def soma_numeros():
    return sum(range(100))

# Resultado em microsegundos
timeit.timeit('soma_numeros()', setup="from __main__ import soma_numeros")
```

```python
# Exemplo básico de medição de tempo de execução usando o magic %timeit do IPython
def soma_numero():
    return sum(range(100))

%timeit soma_numero()
```

Viram a diferença? Com menos linhas de código o magic **%timeit** do IPython consegue nos fornecer muito mais informação que a função **timeit** básica do Python. Além do mais, executa o mesmo códino *n* vezes para extrair uma medida mais precisa.

Além dos *magics* o IPython também fornece a opção de executar *shortcuts* para algumas funcionalidades. Por exemplo, algumas funções do *bash* como *ls* estão disponível por padrão.

```python
ls
```

## 3. Jupyter  

```python
Image('http://jupyter.org/assets/jupyterpreview.png')
```

O Jupyter nasceu de uma funcionalidade antes fornecida pelo IPython, os famosos *notebooks*. É um projeto que integra a web, Python e o IPython permitindo que o ambiente interativo do Ipython seja executado dentro do navegador. 

Aliás, todo esse post foi escrito dentro de um Jupyter notebook, que permite a criação de células em [Markdown](https://pt.wikipedia.org/wiki/Markdown), código Python e funcionalidades da web (HTML, CSS, Javascript). 

O mais interessante dos notebooks é que eles permitem a análise de dados através de anotações descritivas, o que facilita a colaboração entre integrantes de um mesmo time.

Por exemplo, temos a seguinte função linear descrita abaixo:

$$ f(x) = 2x + 0.5 - 4.5x $$ 

O código em Python para essa função é: 

```python
def f(x):
    return 2*x + 0.5 - 4.5*x
```

Usaremos o **matplotlib** para desenhar o gráfico dos número de 0 a 50 para essa função.

```python
# importando a função de plot
import matplotlib.pyplot as plt 
# magic do Ipython que permite que o resultado de um plot seja automaticamente exibido como output
%matplotlib inline 

plt.plot([f(x) for x in range(51)])
```

O Jupyter também permite a utilização de *widgets* HTML para incrementar a interatividade de uma análise. Podemos usar a função $f$ definida mais acima em conjunto com um *slider* e obtermos o resultado em tempo real.

```python
from ipywidgets import interact

interact(f, x=10)
```
